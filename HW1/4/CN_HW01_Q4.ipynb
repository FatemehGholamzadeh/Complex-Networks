```python
G = nx.DiGraph()
i=0
with open (path, 'r') as f:
    for row in csv.reader(f,delimiter='\t'):
      for a in row[1].split(','):
        if(a!=''):
          G.add_edge(int(row[0]),int(a))
        if(a==''):
           G.add_node(int(row[0]))

    
```

```python
print(nx.info(G))
```

```python
 Avg_degree =(G.size())/len(G)
 print("Average Degree is : ",Avg_degree)
```

```python
print(nx.average_clustering(G))
```

```python
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.figure(figsize=(12,8))
    plt.hist(degrees, facecolor='m',bins=200,linewidth=6)
    plt.xlabel("Degree")
    plt.ylabel("Frequency")
    plt.title("Degree Histogram")
    plt.show()

plot_degree_dist(G)
```
